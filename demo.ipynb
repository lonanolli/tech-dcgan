```python
import json
import torch
import matplotlib.pyplot as plt

from src.models import Generator
from src import visualization as viz
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
latent_dim = 100

ckpt = torch.load("outputs/dcgan_weights.pt", map_location=device)
G = Generator(latent_dim).to(device)
G.load_state_dict(ckpt["G"])
G.eval()
```

```python
with open("outputs/losses.json", "r") as f:
    losses = json.load(f)

viz.plot_losses(losses["g_losses"], losses["d_losses"])
plt.show()
```

```python
z = viz.sample_images_grid(G, latent_dim=latent_dim, device=device, n=10, nrow=10)
plt.title("10 Generated Flower Images")
plt.show()
```
